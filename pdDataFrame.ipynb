This is an example on how to create a dataframe in pandas with 4 columns and 100 rows. 
The data is randonmly generated and the column names are assigned as A, B, C, and D

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randint(0,100,size=(100, 4)), columns=list('ABCD'))
```

```python
df.head()
```

```python
# Counting how many values on each row are higher than the values of column D. 
# In this case, 3 is the maximum as there are only three colomns more than D
df['Count'] = df[[ 'A','B','C']].ge(df['D'],axis=0).sum(axis=1)
df
```
