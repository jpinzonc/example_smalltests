```python
import pandas as pd
import numpy as np
```

```python
series = pd.Series(pd.read_clipboard(header = None)[1].values)
```

```python
df = pd.DataFrame({'Series': series})
df
```

```python
quantiles = df['Series'].quantile([0, .25, .5,0.75,1]).to_frame('quantiles').reset_index(drop = True)
quantiles.quantiles
```

```python
for quant, Q in enumerate(['Q1', 'Q2', 'Q3', 'Q4']):
    quant = quant + 1
    df.loc[:,Q] = np.where((df.Series > quantiles.quantiles[quant-1]) & (df.Series <= quantiles.quantiles[quant]), 1, 0)
df
```
