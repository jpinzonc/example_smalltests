```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

date_1 = datetime(2017,3,4).date()
days = pd.date_range(date_1, date_1 + timedelta(days = 200), freq='D')
np.random.seed(seed=1111)
data = np.random.randint(1, high=100, size=len(days))
df = pd.DataFrame({'date': days, 'number': data})
df = df.set_index('date')
display(df)
```

```python
from datetime import datetime, timedelta
import pandas as pd
import numpy as np

date_today = datetime.now().date()
ndays = 200
df = pd.DataFrame({'date': [date_today - timedelta(days=x) for x in range(ndays)], 
                   'number': pd.Series(np.random.randn(ndays))})
df=df.set_index('date')
df
```
